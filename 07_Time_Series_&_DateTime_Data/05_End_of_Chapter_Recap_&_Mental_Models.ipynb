## Topic 2: Resampling Time Series

Resample -> Group data by time frequency + Aggregate  

1. Downsampling:  
    Daily -> Monthly - `df.resample("ME").sum()`
2. Upsampling:  
    Yearly -> Monthly - `df.resample("ME")`  
    But this return object so we add `df.resample("ME").asfreq()` So we get DataFrame.  
    After upsampling we get NaN/Missing values. For best practice dont fill imediatly after upsampling. Fisrt see the data and then after confirming the fill-bfill/ffill do fill.
    To fill it up.
3. Fill missing values
    1. Forward: `upsampled_df.ffill()`
    2. Backward: `upsampled_df.bfill()`
4. Resample with multiple aggregate: `df.resample("D").agg({"Sales": ["sum" ,"min",...]})`

## Topic 3: Shifting

1. `df["col_to_store"] = df["col_to_shift"].shift(1)` - Shift sales down by 1.
2. `df["col_to_store"] = df["col_to_shift"].shift(2)` - Shift sales down by 2.
3. `df[".."].shift(-1)` - Shift row up by 1
4. For DateTime index - Freq based shifting - `df["Col_to_shift"].shift(freq="2D")`

## Topic 4: Time based Slicing

1. Single date - `df.loc["date"]`
2. Slice - `df.loc["Start_date", "end_date"]` --- Both start and end are inclusive.
3. Use partial dates:
    1. Month - `df.loc["2023-01"]` give jan data
    2. year - `df.loc["2023"]` gives full 2023 year
4. Conditions also work on DateTime - `df[df.index > "2023-01-01"]` or `df[() & ()]`
5. Time selection:
    for that we should have time in date - `df.between_time(start : end)` - `df.between_time("09:00", "10:00")`
6. Select specific time - `df.at_time("19:00")`
