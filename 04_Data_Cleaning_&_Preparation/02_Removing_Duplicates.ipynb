# Removing duplicates

`.drop_duplicates()`

```python
# sample data frame
df = pd.DataFrame({
    "Name": ["Onkar", "Amit", "Sara", "Amit", "Sara"],
    "City": ["Pune", "Mumbai", "Nashik", "Mumbai", "Nagpur"],
    "Age": [21, 25, 23, 25, 23]
})

df
```

## 1. Find duplicate rows

`.duplicated()` -> Returns True for that rows which are repeated

```python
df.duplicated()
```

## 2. Count duplicates

`.duplicated().sum()` -> Returns the count of duplicate rows

```python
df.duplicated().sum()
```

## 3. Remove duplicate rows

`.drop_duplicates()` -> Remove duplicate rows and return a clean DataFrame without duplicate rows

```python
df.drop_duplicates()
```

The 3rd row is droped which is duplicate in original df

## 4. Remove duplicates based on specific columns

```python
df.drop_duplicates(subset=["City"]), df.drop_duplicates(subset=["Name"]) 
```

To keep last occurrence use `keep='last'`, First is default.  
Options for keep: 
1. "first" - Default
2. "last" - Keeps last occurrence and drops all above duplicates
3. False - remove all rows which are duplicates

Below ex to understand `keep=False`.  
On the basis on name `keep=False` removed all rows which are duplicates.  
Nither first nor last occurrence is keeped.

```python
df.drop_duplicates(subset=["Name"], keep=False)
```

## 5. Remove all duplicates (keep none)

```python
df.drop_duplicates(keep=False)
```

## 7. Drop duplicates + reset index

```python
df.drop_duplicates().reset_index()
```

```python
df.drop_duplicates().reset_index(drop=True)
# drop=True -> Drops the old index 
```
