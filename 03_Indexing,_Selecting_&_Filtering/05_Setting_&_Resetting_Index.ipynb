# Setting & Resetting Index

`.set_index()` and `.reset_index()`

```python
df = pd.DataFrame({
    "ID": [101, 102, 103, 104],
    "Name": ["Onkar", "Amit", "Sara", "Rohit"],
    "City": ["Pune", "Mumbai", "Nashik", "Pune"],
    "Salary": [50000, 65000, 55000, 70000]
})

df
```

Right now above table have default index which is (0,1,2,....)

## 1. What is an index?

index = row labels. Used to identify the rows.  
It can be number, string, date, IDs, etc.  
It is like primary key of the table.

## 2. `.set_index()` -> Change a column into a index

```python
# access by default index
df.loc[1]
```

```python
df_indexed = df.set_index("ID") # Set ID column as index
df_indexed
```

```python
df_indexed.loc[101] # access rows by using ID. As we set IDcolumn as index
```

```python
# But when setting the index the column is droped from data if you want column stay in the data so doo
df.set_index("ID", drop=False)
```

Points:  
1. `.set_index("col_name")` -> to set the column as index (but the column drops from the data)  
Too keep it do
2. `.set_index("col_name", drop=False)`

## 3. `.reset_index()` -> Resets the index to default

```python
df_indexed.reset_index()
```

```python
df_indexed.reset_index(drop="True")
```

Points:  
1. `.reset_index()` -> It reset the index to default only.
2. `.reset_index(drop=True)` -> It drops the column which is index.

## 4. When should you set an index?

When a column uniquely determines the rows

## 4. When should you set an index?

After filtering rows and before exporting data
