# Understanding Series and DataFrame

```python
import pandas as pd
```

## What is Series?
Series is a single column of data.

```python
s = pd.Series([10, 20, 30, 40])
print(s)
```

### Access particular value

```python
print(s[2])
print(s[1])
```

## What is DataFrames?
Data Frame is a Table (rows + columns). Collection of series.

```python
df = pd.DataFrame({
    "Name" : ["AAA", "BBB", "CCC"],
    "Age" : [10, 20, 30]
})
print(df)
```

### Access particular value

```python
print(df["Age"])
```

## Basic operations

### On Series

```python
print(s.values)
print(s.index)
```

### On DataFrame

```python
print(df.shape)
print("----")
print(df.columns)
print("----")
print(df.index)
print("----")
print(df.dtypes)
```
