```python
import pandas as pd
```

# Summary Statistics

```python
#Sample DataFrame
df = pd.DataFrame({
    "Student": ["Onkar", "Amit", "Sara", "Rohit", "Neha"],
    "Math": [85, 90, 88, 92, 76],
    "Science": [78, 92, 91, 89, 80],
    "English": [80, 86, 89, 90, 75]
})

df
```

## 1. Mean `.mean()`

```python
df["Math"].mean()
```

```python
df.mean(numeric_only=True)
```

## 2. Median `.median()`

```python
df["Math"].median()
```

## 3. Mode `.mode()`

```python
df["Math"].mode()
```

## 4. Quantiles `.quantiles()`

```python
df["Math"].quantile(0.25)     # 25th percentile
```

```python
df["Math"].quantile(0.50)     # same as median
```

```python
df["Math"].quantile(0.75)     # 75th percentile
```

## 5. Variance & Dtandard Deviation `.var()` & `.std()`

```python
df["Math"].var()     # Variance
```

```python
df["Math"].std()     # Standard Deviation
```

## 6. Correlation `.corr()`

```python
df[["Math", "Science"]].corr()
```

```python
df.corr(numeric_only=True)
```

## 7. Covariance `.cov()`

```python
df[["Math", "Science"]].cov()
```

```python
df.cov(numeric_only=True)
```

## 8. `describe()` - One for all

```python
df.describe()
```

## 9. Apply stats column-wise or row-wise

Column-wise is default, for row-wise we have to pass `axis=1`

```python
df.mean(numeric_only=True)     # Column wise
```

```python
df.mean(numeric_only=True, axis=1)     # Row wise
```


---

Summary

1. Mean - `.mean()` -> Mathematical mean
2. Median - `.median()` -> Mathematical median
3. Mode - `.mode()` -> Mathematical mode
4. Quantiles - `.quantiles()` -> Mathematical 
5. Variance - `.var()` -> Mathematical Variance
6. Standard Deviation - `.std()` -> Mathematical SD
7. Correlation - `.corr()` -> Mathematical correlation
8. Covariance - `.cov()` -> Mathematical covatiance
9. One for most - `.describe()` -> This gives count, mean, std, min, 20%, 50%, 75%, max
10. Axis based calculations - `.mean(axis=0/1)` -> For Column-wise/Row-wise. 0: Column (default) and 1: Row-wise
