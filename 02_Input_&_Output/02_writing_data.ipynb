```python
import pandas as pd
```

# Writing Data (Exporting)

## 0. Setup Ex DataFrame

```python
df = pd.DataFrame({
    "Name": ["Aaa", "Bbb", "Ccc"],
    "Marks": [88, 92, 95],
    "City": ["Pune", "Mumbai", "Satara"]
})
df
```

## 1. Write to CSV -> `.to_csv()`

Points: 

1. Write csv file -> `.to_csv("file_name.csv")`
2. Written file take index column also. To remove that -> `.to_csv("file_name.csv", index=False)`
3. If user wants other seperator it can be -> `.to_csv("file_name.csv", index=False, sep=";")`
4. If used wants specific column to write -> `df["col_name"].to_csv("file_name.csv", index=False)`
4. If used wants multiple columns to write -> `df[list-of-col-name].to_csv("file_name.csv", index=False)`

```python
df.to_csv("written.csv") # full table with index are converted to csv
```

```python
df.to_csv("written1.csv", index=False) # excludes the index
```

```python
df.to_csv("written2.csv", index=False, sep=';') # Can add another seperator
```

```python
df["Name"].to_csv("written3.csv", index=False) # Can add specific columns using subscript
```

```python
df[["Name", "City"]].to_csv("written4.csv", index=False) # Can add multiple specific columns just use [[list.of.coluns]]
```

## 2. Write to Excel -> `.to_excel()`

Points:

1. Basic use -> `df.to_excel("file_name.xlsx", index=False)`  
   It creates a excel file "file_name.xlsx" and the sheet name remains default as per excel
2. Write specific sheet -> `df.to_excel("file_name.xlsx", index=False, sheet_name="Score")`  
   It creates a "file_name.xlsx" excel file and inside it creates a "Score" sheet and write in that sheet

```python
df.to_excel("written.xlsx", index=False) # Creates a excel file (default sheet making)
```

```python
df.to_excel("written1.xlsx", index=False, sheet_name="Score") # Creates a sheet named "Score" in a excel file
```

## 3. Write to JSON -> `.to_json()`

Points:

```python
df.to_json("written.json", orient="records") # creates a json file with format which os widely used [{},{}]
```
