````python
import pandas as pd
````

# Pivot Tables & Crosstabulation

````python
# sample dataframe

df = pd.DataFrame({
    "Employee": ["Onkar", "Amit", "Sara", "Rohit", "Neha", "Karan"],
    "Department": ["IT", "IT", "HR", "IT", "HR", "Finance"],
    "Gender": ["M", "M", "F", "M", "F", "M"],
    "Salary": [50000, 65000, 55000, 70000, 48000, 60000]
})

df
````

## 1. What is a pivot table?

**A pivot table:**
- Groups data
- Aggregates values
- Presents results in a matrix form

## 2. Basic pivot table -> `.pivot_table()`

````python
pd.pivot_table(
    df,
    values = "Salary",
    index = "Department",
    aggfunc = "mean"
)
````

**index** -> rows  
**values** -> On which column the agg. fun have to apply  
**aggfunc** -> Aggregate function

## 3. Multiple aggregate functions

````python
pd.pivot_table(
    df,
    index = "Department",
    values = "Salary",
    aggfunc = ["mean", "max", "min", "count"]
)
````

## 4. Pivot with columns (2D analysis)

````python
pd.pivot_table(
    df,
    index = "Department",
    values = "Salary",
    columns = "Gender",
    aggfunc = "mean"
)
````

The table have **index/rows** -> **Department**  
and **column** -> **Gender**

````python
pd.pivot_table(
    df,
    values = "Salary",
    index = "Department",
    aggfunc = "mean",
    columns = ["Gender", "Employee"],
    fill_value = 0
)
````

## 5. Handling missing values in  pivot

````python
pd.pivot_table(
    df,
    values = "Salary",
    index = "Department",
    columns = "Gender",
    aggfunc = "mean",
    fill_value = 0
)
````

## 6. Multiple values in pivot

````python
df["Bonus"] = df["Salary"] * 0.10
````

````python
pd.pivot_table(
    df,
    values = ["Salary", "Bonus"],
    index = "Department",
    aggfunc = "mean"
)
````

````python
pd.pivot_table(
    df,
    values = ["Salary", "Bonus"],
    index = "Department",
    aggfunc = "mean",
    columns = "Gender",
)
````

````python
pd.pivot_table(
    df,
    values = ["Salary", "Bonus"],
    index = "Department",
    aggfunc = "mean",
    columns = "Gender",
    fill_value = 0
)
````

## 7. `pd.crosstab()` â€” Frequency Tables

How many Gender(M or F) are in Departments

````python
pd.crosstab(df["Department"], df["Gender"])
````

````python
pd.crosstab(df["Department"], df["Salary"])
````

## 8. Crosstab with normalization (percentages)

````python
pd.crosstab(
    df["Department"],
    df["Gender"],
    normalize = "index"
)
````

````python
m = pd.crosstab(
    df["Department"],
    df["Gender"],
    normalize = "index"
)
````
